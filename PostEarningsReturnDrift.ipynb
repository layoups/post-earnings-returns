```python
import pandas as pd 
import numpy as np
import copy
```

```python
dsf = pd.read_csv('dsf.csv')
dsf.head()
```

```python
df = dsf[dsf.DATE >= 20180101]
df
```

```python
df_an = copy.deepcopy(df)
df_an = df_an[['PERMNO', 'DATE', 'RET']]
df_an.rename({'DATE': 'ANNDATS'}, inplace=True, axis=1)
df_an
```

```python
df_permno = pd.read_csv('permno_to_tickers.csv')
df_permno
```

```python
df_earnings = pd.read_csv('ActualEarningsData20182022.csv')
df_earnings.drop('TICKER', inplace=True, axis=1)
df_earnings.rename({'OFTIC': 'TICKER'}, inplace=True, axis=1)
df_earnings
```

```python
df_pe = pd.merge(df_permno, df_earnings, on='TICKER', how='left')
df_pe = df_pe[['TICKER', 'PERMNO', 'ANNDATS']]
df_pe.dropna(inplace=True)
```

```python
df_merged = pd.merge(df_pe, df_an, on=['PERMNO', 'ANNDATS'], how='left')
df_merged
```
